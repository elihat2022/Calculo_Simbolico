## Cálculo de Límites usando Sympy

Como función de ejemplo para el cálculo de limites se procede a usar la siguiente funcion:
$f(x)=\frac{1}{x+1}$


```python
import sympy 
from sympy.plotting import plot

x = sympy.symbols('x')
f = 1/(x+1)
plot(f, xlim = [-4,2], ylim=[-4,4])
```

```python
# Calculo de Limites usando Sympy cuando tiende a infinito
limite = sympy.Limit(f, x, sympy.S.Infinity)
limite_resultado = limite.doit()
print("El limite de f cuando x tiende a infinito es:", limite_resultado)
```

```python
# Calculo de Limites usando Sympy cuando tiende a -1
limite = sympy.Limit(f, x, -1)
limite_resultado = limite.doit()
print("El limite de f cuando x tiende a -1 es:", limite_resultado)
```

```python
# Calculo de Limites usando Sympy cuando tiene direccion izquierda en -1
limite = sympy.Limit(f, x, -1, dir='-')
limite_resultado = limite.doit()
print("El limite de f cuando x tiende a -1 por la izquierda es:", limite_resultado)
```

## Cálculo de derivadas usando Sympy

Se procede a usar la siguiente funcion para el calculo de derivadas:
$f(x)=\frac{3x^3 + 17x^2 + 6x + 1}{2x^3 -x +3}$

```python
x = sympy.symbols('x')
fp = (3*x**3+17*x**2+6*x+1)/(2*x**3 - x +3)
plot(fp, xlim = [-10,10], ylim=[-10,10])
```

```python
# Cálculo de derivadas usando Sympy
derivada = sympy.Derivative(fp, x)
derivada_resultado = derivada.doit()
print("La derivada de f es:", derivada_resultado)
```

```python
soluciones = sympy.solve(fp, x)
print("Las soluciones de f son:", soluciones)
[complex(sol) for sol in soluciones]
```

```python
# Cálculo de la segunda derivada a partir de la función fp ya definida
segunda_derivada = sympy.diff(fp, x, 2)
segunda_derivada_simplificada = sympy.simplify(segunda_derivada)

print("Segunda derivada (sin simplificar):", segunda_derivada)
print("Segunda derivada (simplificada):", segunda_derivada_simplificada)
```
